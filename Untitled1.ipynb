```python
import numpy as np
import pandas as pd

from datasetsforecast.hierarchical import HierarchicalData


from mlforecast import MLForecast
from window_ops.expanding import expanding_mean
from mlforecast.utils import PredictionIntervals
from mlforecast.target_transforms import Differences
import xgboost as xgb

from hierarchicalforecast.methods import BottomUp, MinTrace, TopDown
from hierarchicalforecast.utils import HierarchicalPlot
from hierarchicalforecast.core import HierarchicalReconciliation
```

```python
Y_df = pd.read_csv('https://raw.githubusercontent.com/Nixtla/transfer-learning-time-series/main/datasets/tourism.csv')
Y_df = Y_df.rename({'Trips': 'y', 'Quarter': 'ds'}, axis=1)
Y_df.insert(0, 'Country', 'Australia')
Y_df = Y_df[['Country', 'Region', 'State', 'Purpose', 'ds', 'y']]
Y_df['ds'] = Y_df['ds'].str.replace(r'(\d+) (Q\d)', r'\1-\2', regex=True)
Y_df['ds'] = pd.to_datetime(Y_df['ds'])
Y_df.head()
```

```python
spec = [
    ['Country'],
    ['Country', 'State'],
    ['Country', 'State',"Purpose"]
]
```

```python
from hierarchicalforecast.utils import aggregate
```

```python
Y_df, S_df, tags = aggregate(Y_df, spec) 
#Y_df = train data, S_df = summing matrix, tags = disaggregated data (hierarchical aggregation indexes )
Y_df = Y_df.reset_index()
```

```python
Y_test_df = Y_df.groupby('unique_id').tail(8)  #get last 8 data(last 8 quarters-2 years) of every tag (every aggregated data)
Y_train_df = Y_df.drop(Y_test_df.index)
```

```python
Y_df.groupby('unique_id').size()
```

```python
Y_train_df.groupby('unique_id').size()
```

```python
mf = MLForecast(models=[xgb.XGBRegressor()], 
                freq='QS',  #Pandas offset or integer denoting the frequency of the series. QS = Quarter Start Frequency
                lags=[1,3,12,24], #t-1
                )
```

 <h5>Lags</h5> <br> Lags of the target to use as features. 

```python
mf.fit(Y_train_df)
```

```python
horizon = 8  #Number of periods to predict. We predict last 8 quarters.
level = np.arange(0, 100, 2) #Confidence levels between 0 and 100 for prediction intervals. We dont use for point estimation.
Y_hat_mf = mf.predict(horizon).set_index('unique_id')
```

```python
Y_hat_mf
```

```python
reconcilers = [
    BottomUp(),
    MinTrace('ols')
]
hrec = HierarchicalReconciliation(reconcilers=reconcilers) # which method is used for Reconciliation
```

```python
Y_rec_df = hrec.reconcile(Y_hat_df=Y_hat_mf, Y_df=Y_train_df, S=S_df, tags=tags) 
```

 The `reconcile` method is analogous to SKLearn `fit_predict` method, it applies different reconciliation techniques instantiated in the `reconcilers` list. 
 Y_hat_df = XGBoost forecasting results (last 2 years)
 Y_df = Train data

```python
from hierarchicalforecast.evaluation import HierarchicalEvaluation
```

```python
def rmse(y, y_hat):
    return np.mean(np.sqrt(np.mean((y-y_hat)**2, axis=1)))

def mase(y, y_hat, y_insample, seasonality=4):
    errors = np.mean(np.abs(y - y_hat), axis=1)
    scale = np.mean(np.abs(y_insample[:, seasonality:] - y_insample[:, :-seasonality]), axis=1)
    return np.mean(errors / scale)

eval_tags = {}
eval_tags['Total'] = tags['Country']
eval_tags['State'] = tags['Country/State']
eval_tags['Purpose'] = tags['Country/State/Purpose']

eval_tags['All'] = np.concatenate(list(tags.values()))

evaluator = HierarchicalEvaluation(evaluators=[rmse, mase])
evaluation = evaluator.evaluate(
        Y_hat_df=Y_rec_df, Y_test_df=Y_test_df.set_index('unique_id'),
        tags=eval_tags, Y_df=Y_train_df.set_index('unique_id')
)
evaluation = evaluation.drop('Overall')
evaluation.columns = ['Base', 'BottomUp','MinTrace(ols)']
evaluation = evaluation.applymap('{:.2f}'.format)
```
